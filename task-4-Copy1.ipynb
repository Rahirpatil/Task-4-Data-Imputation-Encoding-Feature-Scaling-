```python
#to fill the empty values with mean
#SimpleImputer used for find simple mean
#fit_transform used to fill values
from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values = np.nan, strategy= 'mean')
ds[["Age","Salary"]]= si.fit_transform(ds[["Age","Salary"]])
print(ds)
```

```python
#Encoding( OneHotEncoding)
#sparse_output used to output representation in binary 
from sklearn.preprocessing import OneHotEncoder
ohe= OneHotEncoder(sparse_output= False)
encoded_data = ohe.fit_transform(ds[["Country"]])
encoded_data
```

```python
ds["Country"].unique()
```

```python
#get_feature_names_out used to add names of given column
column_names= ohe.get_feature_names_out(["Country"])
encoded_df=pd.DataFrame(encoded_data, columns = column_names)
encoded_df
```

```python
#join it to original dataset
ds=pd.concat([ds,encoded_df],axis=1)
print(ds)
```

```python
#Drop previous columns which are now not necessary
ds=ds.drop(columns=["Country"])
print(ds)
```

```python
#labelEncoder encode target labels with value between 0 and 1_classes-1
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds["Purchased"] = le.fit_transform(ds["Purchased"])
ds
```

```python
# OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

# Sample data: Categorical features with an inherent order
data = [['Medium'],['High'], ['Low'], ['Medium'], ['High'], ['Medium'], ['Low'], ['Low']]

# Initialize the OrdinalEncoder
encoder = OrdinalEncoder(categories=[[ 'High', 'Medium', 'Low']])  # Define the order explicitly

# Fit and transform the data
encoded_data = encoder.fit_transform(data)

# Display the encoded data
print("Original Data:", data)
print("Encoded Data:\n", encoded_data)
```

```python
#Feature scaling
#StandardScaler - standerdize features by removing the mean and scaling to unit variance
#Uses formula: z = (x-u) / s
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
ds[["Age","Salary"]] = sc.fit_transform(ds[["Age","Salary"]])
ds
```

```python
# sc.scale_ tells us the standard deviation of each feature in the original data. 
# These values are used by StandardScaler to scale each value to unit variance.
sc.scale_
```
